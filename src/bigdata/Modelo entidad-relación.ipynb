# Modelo Entidad Relación

El esquema se centra en la tabla Persona con llave primaria `id_persona` (PK `id_persona`), que almacena datos demográficos y métricas de salud (sexo, edad, raza, IMC, horas de sueño, salud general y días con afectación física/mental, y dificultad para caminar). 

A partir de esta entidad se derivan tres tablas satélite: 

### Hábitos 
 (PK `id_habito`, FK `id_persona`) con información de consumo de tabaco y alcohol, actividad física y diabetes; 
 
### CondicionesMédicas 
 
(PK `id_condicion`, FK `id_persona`) con antecedentes de accidente cerebrovascular, asma, enfermedad renal y cáncer de piel; y 
 
 
### SaludCardíaca 
 
(PK `id_saludcardiaca`, FK `id_persona`) que registra la variable objetivo `HeartDisease` (enfermedades cardiacas). 

Las relaciones 1:1 entre Persona y cada tabla aseguran, en este diseño, un único registro de hábitos, condiciones y estado cardiaco por persona. Los tipos de datos (`INT`, `VARCHAR`, `FLOAT`) y las claves foráneas garantizan integridad referencial y consistencia.
